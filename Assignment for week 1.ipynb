ASSIGNMENT 1. create a python program that stimulate a game of rock , . the paper, scissors between the user and the computer. the program must define the possible choice, generate a random choice for the computer, determine the winner based on the rules of the rock, paper and scissors

```python
import random

# Step 1: Define possible choices
choices = ["rock", "paper", "scissors"]

# Step 2: Ask the user to make a choice
user_choice = input("Enter your choice (rock, paper, or scissors): ").lower()

# Step 3: Validate user input
if user_choice not in choices:
    print("Invalid choice. Please choose rock, paper, or scissors.")
else:
    # Step 4: Computer makes a random choice
    computer_choice = random.choice(choices)
    print(f"Computer chose: {computer_choice}")

    # Step 5: Determine the winner
    if user_choice == computer_choice:
        print("It's a tie!")
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        print("You win!")
    else:
        print("Computer wins!")
```

winning rules implemented in the game.

| User's Choice | Computer's Choice | Result        |
| ------------- | ----------------- | ------------- |
| rock          | scissors          | User wins     |
| paper         | rock              | User wins     |
| scissors      | paper             | User wins     |
| rock          | paper             | Computer wins |
| paper         | scissors          | Computer wins |
| scissors      | rock              | Computer wins |
| Same choice   | Same choice       | It's a tie    |


ASSIGNMENT 2. create a python program that stimulatesa simple AI agent that can learn to play tic-tac-toe. Tyhe program must definethe game board, check if a playerhas won ,check if the game has won , check if the game is a tie, main game loop, call the main game loop.

```python
import random

# Define the game board
board = [" " for _ in range(9)]

# Print the board
def print_board():
    for i in range(3):
        print("|".join(board[i*3:(i+1)*3]))
        if i < 2:
            print("-----")

# Check if a player has won
def check_win(player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # columns
        [0, 4, 8], [2, 4, 6]              # diagonals
    ]
    return any(all(board[i] == player for i in condition) for condition in win_conditions)

# Check if the game is a tie
def check_tie():
    return all(cell != " " for cell in board)

# Get human player's move
def player_move():
    while True:
        try:
            move = int(input("Enter your move (1-9): ")) - 1
            if board[move] == " ":
                board[move] = "X"
                break
            else:
                print("Cell is already taken!")
        except (ValueError, IndexError):
            print("Invalid input. Please choose a number between 1 and 9.")

# Get AI move (random for simplicity)
def ai_move():
    available_moves = [i for i, cell in enumerate(board) if cell == " "]
    move = random.choice(available_moves)
    board[move] = "O"
    print(f"AI chose position {move + 1}")

# Main game loop
def play_game():
    print("Welcome to Tic-Tac-Toe! You are X, AI is O.")
    print_board()

    while True:
        # Player's move
        player_move()
        print_board()
        
        # Check for player's win or tie after their move
        if check_win("X"):
            print("You win!")
            break
        if check_tie():
            print("It's a tie!")
            break

        # AI's move
        ai_move()
        print_board()
        
        # Check for AI's win or tie after its move
        if check_win("O"):
            print("AI wins!")
            break
        if check_tie():
            print("It's a tie!")
            break

# Call the main game loop
if __name__ == "__main__":
    play_game()
```

AAIGNMENT 3a. read total profit of all the monthsand show it using a line plot 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\User\Documents\company_sales_data.csv")  # Replace with your actual file path

# Ensure the 'month_number' column is an integer
df['month_number'] = df['month_number'].astype(int)

# Assuming the 'total_profit' column contains the profit values
monthly_profit = df.groupby('month_number')['total_profit'].sum()

# Plotting the total profit by month
plt.figure(figsize=(10, 6))
plt.plot(monthly_profit.index, monthly_profit.values, marker='o')
plt.title('Total Profit per Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.xticks(monthly_profit.index)  # Ensures month numbers (1-12) are on x-axis
plt.grid(True)
plt.show()

```

3b. read the data for bathing soap and facewash  of all months and display it using the subplot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (adjust the file path accordingly)
# Assuming the data is in a CSV format and the columns are 'Month_Number', 'BathingSoap', 'Facewash'
data = pd.read_csv(r"C:\Users\User\Documents\company_sales_data.csv")


# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Plotting Bathing Soap data on the first subplot
axes[0].plot(data['month_number'], data['bathingsoap'], marker='o', color='b', label='Bathing Soap')
axes[0].set_title('Bathing Soap Sales over Months')
axes[0].set_xlabel('Month Number')
axes[0].set_ylabel('Sales')
axes[0].grid(True)
axes[0].legend()

# Plotting Facewash data on the second subplot
axes[1].plot(data['month_number'], data['facewash'], marker='s', color='g', label='Facewash')
axes[1].set_title('Facewash Sales over Months')
axes[1].set_xlabel('Month Number')
axes[1].set_ylabel('Sales')
axes[1].grid(True)
axes[1].legend()

# Display the plot
plt.tight_layout()
plt.show()

```
